```python
# @title Default title text
import cv2
from ultralytics import YOLO
import numpy as np
from collections import defaultdict
import os
```

```python
# @title Default title text
# Set your dataset path here
video_path = "/content/drive/MyDrive/Vehicle Detection/Vehicle Dataset/archive (1)/Sample_Video_HighQuality.mp4"  # <-- Change this to your actual video file path

# Optional: Upload video if using Google Colab interactively
# from google.colab import files
# uploaded = files.upload()
# video_path = list(uploaded.keys())[0]
```

```python
# @title Default title text
# Load YOLOv8 model (n = nano, fast and small)
model = YOLO("yolov8n.pt")  # or yolov8s.pt, yolov8m.pt

# Classes from COCO dataset considered as vehicles
vehicle_classes = ['car', 'motorbike', 'bus', 'truck']
```

```python
# @title Default title text
# Open video file
cap = cv2.VideoCapture(video_path)
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

# Create output video writer
out = cv2.VideoWriter("output_vehicle_count.avi",
                      cv2.VideoWriter_fourcc(*'XVID'),
                      20.0, (frame_width, frame_height))

# Counters
total_counts = defaultdict(int)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)[0]
    vehicle_count = 0

    for r in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = r
        class_name = model.names[int(class_id)]
        if class_name in vehicle_classes and score > 0.5:
            vehicle_count += 1
            total_counts[class_name] += 1
            # Draw bounding box and label
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            cv2.putText(frame, class_name, (int(x1), int(y1)-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    # Show total on frame
    cv2.putText(frame, f"Vehicles: {vehicle_count}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
# @title Default title text
print("Vehicle counts by type:")
for vehicle, count in total_counts.items():
    print(f"{vehicle}: {count}")
```
